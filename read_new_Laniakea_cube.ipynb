```python
import numpy as np
import matplotlib.pyplot as plt
import csv
```

```python
def plot_laniakea_cube(cube):
    indices = np.argwhere(cube == 1)
    xs = indices[:, 0]
    ys = indices[:, 1]
    zs = indices[:, 2]

    fig = plt.figure()
    ax = fig.add_subplot(111, projection='3d')
    ax.set_xlabel('X')
    ax.set_ylabel('Y')
    ax.set_zlabel('Z')

    ax.scatter(xs, ys, zs, s=1)

    plt.show()

def save_laniakea_csv(cube, filename):
    indices = np.argwhere(cube == 1)
    data = indices.astype(str)
    header = ['X', 'Y', 'Z']

    with open(filename, 'w', newline='') as csvfile:
        writer = csv.writer(csvfile)
        writer.writerow(header)
        writer.writerows(data)
```

```python
cube = np.load('CF4_new_128-z008_watershed_laniakea.npy')
```

```python
plot_laniakea_cube(cube)
```

```python
save_laniakea_csv(cube, 'laniakea_cells.csv')
```
